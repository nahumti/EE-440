```python
#import cell
import numpy as np
import cv2
from skimage import color
import matplotlib.pyplot as plt
```

```python
#creating bins based on gray scale values
image = cv2.imread('2_1.bmp')
one_band = np.zeros((512, 512))

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        one_band[i, j] = image[i, j, 1]
print(one_band)        
i2 = one_band.flatten()
bins = np.zeros(256)

for i in range(len(i2)):
    bins[int(i2[i])] += 1

plt.stem(bins)
plt.show()

#create pdf
pdf = np.zeros(256)
for i in range(len(bins)):
    pdf[i] = bins[i]/(one_band.shape[0] * one_band.shape[1])
    
plt.stem(pdf)
plt.show()

#check to see if sum to 1
count = 0.0
for i in range(len(pdf)):
    count += pdf[i]
    
print(count)
```
